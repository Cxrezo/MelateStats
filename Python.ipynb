```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#### IMPORTAMOS EL ARCHIVO CSV DE LOS RESULTADOS DEL MELATE
dfresultados = pd.read_csv('Melate.csv')
dfresultados.head()
```

```python
dfR = dfresultados.drop(['NPRODUCTO', 'CONCURSO', 'BOLSA', 'FECHA'], axis=1)
dfR
```

```python
FeqR1 = pd.DataFrame(dfR['R1'].value_counts())
FeqR2 = pd.DataFrame(dfR['R2'].value_counts())
FeqR3 = pd.DataFrame(dfR['R3'].value_counts())
FeqR4 = pd.DataFrame(dfR['R4'].value_counts())
FeqR5 = pd.DataFrame(dfR['R5'].value_counts())
FeqR6 = pd.DataFrame(dfR['R6'].value_counts())
FeqR7 = pd.DataFrame(dfR['R7'].value_counts())
```

```python
FeqR1.reset_index(names= 'Número', inplace= True)
FeqR2.reset_index(names= 'Número', inplace= True)
FeqR3.reset_index(names= 'Número', inplace= True)
FeqR4.reset_index(names= 'Número', inplace= True)
FeqR5.reset_index(names= 'Número', inplace= True)
FeqR6.reset_index(names= 'Número', inplace= True)
FeqR7.reset_index(names= 'Número', inplace= True)
```

```python
FeqR1.sort_values(by = 'Número', inplace= True)
FeqR2.sort_values(by = 'Número', inplace= True)
FeqR3.sort_values(by = 'Número', inplace= True)
FeqR4.sort_values(by = 'Número', inplace= True)
FeqR5.sort_values(by = 'Número', inplace= True)
FeqR6.sort_values(by = 'Número', inplace= True)
FeqR7.sort_values(by = 'Número', inplace= True)
```

```python
FreqJoin = pd.merge_ordered(FeqR1, FeqR2, on = 'Número')
FreqJoin = pd.merge_ordered(FreqJoin, FeqR3, on = 'Número')
FreqJoin = pd.merge_ordered(FreqJoin, FeqR4, on = 'Número')
FreqJoin = pd.merge_ordered(FreqJoin, FeqR5, on = 'Número')
FreqJoin = pd.merge_ordered(FreqJoin, FeqR6, on = 'Número')
FreqJoin = pd.merge_ordered(FreqJoin, FeqR7, on = 'Número')
```

```python
FreqJoin.set_index('Número')
FreqJoin['Número'] = FreqJoin['Número'].astype(str)
```

```python
FreqJoin['Suma'] = FreqJoin.sum(axis=1)
```

```python
FreqJoin.set_index('Número')
dfsum = FreqJoin.drop(['R1', 'R2', 'R3', 'R4', 'R5', 'R6', 'R7'], axis = 1)
dfsum.set_index('Número', inplace= True)
```

```python
plt.gcf().clear()
dfsum.plot(kind = 'bar', color = 'green', y = 'Suma', rot = 90)
plt.xticks(fontsize = 5)
plt.show
```

```python
dfsum.sort_values(by = 'Suma', inplace= True, ascending= False)
dfsum
```
